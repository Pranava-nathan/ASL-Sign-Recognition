```python
import os, glob, warnings, logging
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from IPython.display import display
from graphviz import Source
from dowhy import CausalModel
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import roc_auc_score
from econml.dml import CausalForestDML

# Clean output
warnings.filterwarnings("ignore")
logging.getLogger("dowhy").setLevel(logging.ERROR)
pd.set_option("display.max_columns", 200)
sns.set_theme(style="whitegrid", context="notebook")

RANDOM_STATE = 7
np.random.seed(RANDOM_STATE)
```

```python
# Set path to CSV (auto-detect common names, else use your explicit path)
candidates = glob.glob("./*diabetes*BRFSS2015*.csv") + glob.glob("./*diabetes*health*indicators*.csv")
CSV_PATH = candidates[0] if candidates else "/content/diabetes_binary_health_indicators_BRFSS2015.csv"
assert os.path.exists(CSV_PATH), f"CSV not found at {CSV_PATH}"

df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()

print("Loaded shape:", df.shape)
display(df.head(3))
```

```python
def find_col(cols, tokens):
    for c in cols:
        if all(t in c for t in tokens):
            return c
    return None

cols = df.columns
colmap = {
    "diabetes":  find_col(cols, ["diab"]),
    "exercise":  find_col(cols, ["phys","acti"]),
    "bmi":       find_col(cols, ["bmi"]),
    "sex":       find_col(cols, ["sex"]),
    "age":       find_col(cols, ["age"]),
    "education": find_col(cols, ["educ"]),
    "income":    find_col(cols, ["inco"]),
    "smoker":    find_col(cols, ["smok"]),
    "fruits":    find_col(cols, ["fruit"]),
    "veggies":   find_col(cols, ["vegg"]),
    "alcohol":   find_col(cols, ["hvy","alco"]) or find_col(cols, ["alco"]),
    # optional extras
    "genhlth":   find_col(cols, ["genhl"]),
    "physhlth":  find_col(cols, ["physhl"]),
    "menthlth":  find_col(cols, ["menthl"]),
    "highbp":    find_col(cols, ["highbp"]),
    "highchol":  find_col(cols, ["highchol"]),
    "cholcheck": find_col(cols, ["cholcheck"]),
    "heartdis":  find_col(cols, ["heart"]),
    "stroke":    find_col(cols, ["stroke"]),
    "diffwalk":  find_col(cols, ["diffwalk"]),
}
required = ["diabetes","exercise","age","sex","education","income","smoker","fruits","veggies"]
missing = [k for k in required if colmap.get(k) is None]
assert not missing, f"Missing required columns: {missing}"

d = df.rename(columns={v:k for k,v in colmap.items() if v is not None})

keep = [c for c in [
    "diabetes","exercise","age","sex","education","income","smoker","fruits","veggies","alcohol","bmi",
    "genhlth","physhlth","menthlth","highbp","highchol","cholcheck","heartdis","stroke","diffwalk"
] if c in d.columns]
d = d[keep].dropna()
d = d.apply(pd.to_numeric, errors="coerce").dropna().reset_index(drop=True)

print(f"Tidy shape: {d.shape}")
print(f"Base rates: diabetes = {d['diabetes'].mean():.4f} | exercise = {d['exercise'].mean():.4f}")
```

```python
T = "exercise"
Y = "diabetes"
Z_total = [c for c in ["age","sex","education","income","smoker","fruits","veggies","alcohol"] if c in d.columns]
M = "bmi" if "bmi" in d.columns else None

print("Treatment (T):", T)
print("Outcome   (Y):", Y)
print("Confounders (Z):", Z_total)
print("Mediator (exclude for total effect):", M)
```

```python
dag = f"""
digraph G {{
  rankdir=LR;
  graph [splines=true, overlap=false, fontsize=12, labelloc="t", label="Exercise -> Diabetes (Total Effect)"];
  node  [shape=box, style="filled,rounded", fontname="Helvetica"];

  subgraph cluster_Z {{
    label="Confounders (adjustment set Z)";
    color="#10b981";
    {";".join(Z_total)};
    {''.join([f'{z} [fillcolor="#d1fae5"];\n' for z in Z_total])}
  }}

  subgraph cluster_T {{
    label="Treatment";
    color="#3b82f6";
    exercise [fillcolor="#dbeafe", shape=box];
  }}

  subgraph cluster_Y {{
    label="Outcome";
    color="#ef4444";
    diabetes [fillcolor="#fee2e2", shape=box];
  }}

  subgraph cluster_M {{
    label="Mediator (exclude for total effect)";
    color="#f59e0b";
    {"bmi" if M else ""} [fillcolor="#fef3c7", shape=box];
  }}

  U [label="U (unobserved health preference)", shape=ellipse, fillcolor="#e5e7eb", color="#9ca3af"];

  // Edges
  {";".join([f"{z} -> exercise; {z} -> diabetes" for z in Z_total])};
  exercise -> diabetes;
  {"exercise -> bmi; bmi -> diabetes;" if M else ""}
  U -> exercise; U -> diabetes;
}}
"""
display(Source(dag))
```

```python
data_for_causal = d[[Y, T] + Z_total + ([M] if M else [])].copy()

naive = data_for_causal.loc[data_for_causal[T]==1, Y].mean() - \
        data_for_causal.loc[data_for_causal[T]==0, Y].mean()

model_total = CausalModel(
    data=data_for_causal,
    treatment=T,
    outcome=Y,
    common_causes=Z_total,
    proceed_when_unidentifiable=True
)
identified = model_total.identify_effect()

est_ips = model_total.estimate_effect(identified, method_name="backdoor.propensity_score_weighting",
                                      method_params={"weighting_scheme":"ips_weight"})
est_reg = model_total.estimate_effect(identified, method_name="backdoor.linear_regression")
est_match = model_total.estimate_effect(identified, method_name="backdoor.propensity_score_matching")

# Compact summary table
summary = pd.DataFrame({
    "Estimator": ["Naive diff", "DoWhy IPS", "DoWhy OLS", "DoWhy Matching"],
    "ATE": [naive, est_ips.value, est_reg.value, est_match.value]
})
display(summary.style.format({"ATE": "{:.4f}"}))
```

```python
print("Refuters (on IPS estimate):")
print(model_total.refute_estimate(identified, est_ips, method_name="random_common_cause"))
print(model_total.refute_estimate(identified, est_ips, method_name="placebo_treatment_refuter"))
print(model_total.refute_estimate(identified, est_ips, method_name="data_subset_refuter"))
```

```python
X = data_for_causal[Z_total].values
T_arr = data_for_causal[T].astype(int).values
Y_arr = data_for_causal[Y].astype(float).values

cf = CausalForestDML(
    discrete_treatment=True,
    random_state=RANDOM_STATE,
    n_estimators=400,
    min_samples_leaf=40,
    model_t=RandomForestClassifier(n_estimators=200, min_samples_leaf=20, random_state=RANDOM_STATE),
    model_y=RandomForestRegressor(n_estimators=200, min_samples_leaf=20, random_state=RANDOM_STATE),
)
cf.fit(Y_arr, T_arr, X=X)

ate_cf = float(cf.ate(X))
try:
    lb, ub = cf.ate_interval(X, alpha=0.05)
    lb, ub = float(lb), float(ub)
except Exception:
    inf = cf.ate_inference(X)
    if hasattr(inf, "summary_frame"):
        sf = inf.summary_frame(alpha=0.05)
        lb, ub = float(sf["lower"].iloc[0]), float(sf["upper"].iloc[0])
    else:
        lb, ub = np.nan, np.nan

print(f"EconML CausalForest ATE: {ate_cf:.4f} | 95% CI [{lb:.4f}, {ub:.4f}]")

# CATE by age (if available)
if "age" in data_for_causal.columns:
    cate = cf.effect(X)
    tmp = data_for_causal[["age"]].copy()
    tmp["cate"] = cate
    age_grp = tmp.groupby("age")["cate"].agg(["count","mean"]).reset_index()

    fig, ax = plt.subplots(figsize=(6,3.8))
    sns.lineplot(data=age_grp, x="age", y="mean", marker="o", ax=ax)
    ax.set(title="CATE by Age (effect of exercise on diabetes)", xlabel="Age category", ylabel="Effect (risk difference)")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

def iptw_weights(T, Z_df):
    lr = LogisticRegression(max_iter=1000)
    lr.fit(Z_df, T)
    ps = lr.predict_proba(Z_df)[:,1]
    p_t = T.mean()
    w = np.where(T==1, p_t/ps, (1-p_t)/(1-ps))
    return w, ps

def smd_unweighted(x, t):
    m1, m0 = x[t==1].mean(), x[t==0].mean()
    v1, v0 = x[t==1].var(ddof=1), x[t==0].var(ddof=1)
    return (m1-m0)/np.sqrt((v1+v0)/2 + 1e-9)

def smd_weighted(x, t, w):
    w1, w0 = w[t==1], w[t==0]
    x1, x0 = x[t==1], x[t==0]
    m1 = np.average(x1, weights=w1); m0 = np.average(x0, weights=w0)
    v1 = np.average((x1-m1)**2, weights=w1); v0 = np.average((x0-m0)**2, weights=w0)
    return (m1-m0)/np.sqrt((v1+v0)/2 + 1e-9)

Z_df = data_for_causal[Z_total].astype(float)
T_vec = data_for_causal[T].astype(int).values
w, ps = iptw_weights(T_vec, Z_df)

balance = []
for c in Z_total:
    x = Z_df[c].values
    balance.append([c, smd_unweighted(x, T_vec), smd_weighted(x, T_vec, w)])
balance = pd.DataFrame(balance, columns=["variable","SMD_before","SMD_after"]).sort_values("SMD_before", key=lambda s: s.abs(), ascending=False)
display(balance.style.format({"SMD_before": "{:.3f}", "SMD_after": "{:.3f}"}))

fig, ax = plt.subplots(figsize=(7, 0.4*len(balance)+1))
sns.scatterplot(data=balance, x="SMD_before", y="variable", label="Before", s=50, ax=ax)
sns.scatterplot(data=balance, x="SMD_after",  y="variable", label="After IPTW", s=50, ax=ax)
ax.axvline(0.1, color="gray", linestyle="--", alpha=0.6); ax.axvline(-0.1, color="gray", linestyle="--", alpha=0.6)
ax.set(title="Covariate Balance (Standardized Mean Difference)", xlabel="SMD", ylabel="")
plt.tight_layout()
plt.show()

print("Propensity AUC (sanity check, not a causal metric):", roc_auc_score(T_vec, ps))
```

```python
if M:
    Z_direct = Z_total + [M]
    model_direct = CausalModel(
        data=data_for_causal,
        treatment=T,
        outcome=Y,
        common_causes=Z_direct,
        proceed_when_unidentifiable=True
    )
    ident_direct = model_direct.identify_effect()
    est_ips_direct = model_direct.estimate_effect(ident_direct, method_name="backdoor.propensity_score_weighting",
                                                  method_params={"weighting_scheme":"ips_weight"})
    print(f"Direct effect (IPS, adjusted for BMI): {est_ips_direct.value:.4f}")
```

```python
rows = [
    ("Naive difference",        float(naive),       np.nan,  np.nan),
    ("DoWhy ATE (IPS)",         float(est_ips.value), np.nan,  np.nan),
    ("DoWhy ATE (OLS)",         float(est_reg.value), np.nan,  np.nan),
    ("DoWhy ATE (Matching)",    float(est_match.value), np.nan, np.nan),
    ("EconML CausalForest ATE", float(ate_cf),      float(lb), float(ub)),
]
if M and "est_ips_direct" in locals():
    rows.append(("Direct effect (IPS, +BMI)", float(est_ips_direct.value), np.nan, np.nan))

final = pd.DataFrame(rows, columns=["Estimate","Point","CI_lower","CI_upper"])
display(final.style.format({"Point":"{:.4f}","CI_lower":"{:.4f}","CI_upper":"{:.4f}"}))
```
